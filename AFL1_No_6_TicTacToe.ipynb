```python
import numpy as np
import copy as cp

# 6a. The player should be able to choose the character (X or O). For initialization X will play first. 
def game_initializer():
  init_board = [['0', '1', '2'], 
                ['3', '4', '5'],
                ['6', '7', '8']]
  choice = input("\nSelect your sign (X or O)\nYour Choice: ")
  if choice == 'O' or choice == 'X':
    game(choice, turn = 1, board = init_board)
  else:
    print("\nPlease enter 'X' or 'O' and try again.")
    game_initializer()

# 6b. The action should return set of possible action. You can represent each action as a tuple (x,y) where x corresponds to the row and y corresponds to the cell. Possible move are any cells that do not have X or O.
# Iterasi angka sesuai di board yang dihubungkan dengan end untuk jarak spasi
def board_display(board):
  print('\nTictactoe Board:\n\n' + '\n\n'.join(['  '.join(x) for x in board]))

# 6c. The result function takes a board and an action as input and should return a new board state without modifying the original board.
# Membuat board baru tanpa memodifikasi board aslinya
def check_movement(board, action, sign):
  new_board = cp.deepcopy(board)

  i = 2
  if action <= 2:
    i = 0
  elif action >= 3 and action <= 5:
    i = 1
    
  location = [i,(action-(i*3))]

  if new_board[location[0]][location[1]] == str(action):
    new_board[location[0]][location[1]] = sign
    return new_board

# 6c (cont'd). The winner function should accept a board as input and return winner of the board if there is one.
# Cek posisinya yang terisi secara horizontal
def check_horizontal(board):
  # Jika value unique nya hanya ada 1, berarti benar
  unique_row = [np.unique(row) for row in board]
  len_unique = np.array([len(unique) for unique in unique_row])
  # Cek jika panjangnya sama dengan 1, berarti sudah ditemukan pemenangnya
  if any(len_unique == 1):
    # Membalikkan index dari angka terkecil
    return unique_row[np.argmin(len_unique)][0]

# Cek posisinya yang terisi secara horizontal
# Melakukan transpose Array
def check_vertical(board):
  return check_horizontal(np.array(board).T)

# Cek posisinya yang terisi secara diagonal
def check_diagonal(board):
  diagonals = [[board[0][0], board[1][1], board[2][2]], [board[0][2], board[1][1], board[2][0]]]
  return check_horizontal(diagonals)

# Membalikkan siapa yang menjadi pemenangnya hanya jika kondisinya True
def check_win(board):
  if bool(check_horizontal(board)):
    return check_horizontal(board)
  if bool(check_vertical(board)):
    return check_vertical(board)
  if bool(check_diagonal(board)):
    return check_diagonal(board)
  return False

# 6d. The terminal function should accept a board as input and return a Boolean value indicating whether the game is over. 
# Membalikkan value hanya jika permainan berakhir seri.
def check_tie(board):
  flat_board = board[0] + board[1] + board[2] 

  return len(np.unique(flat_board)) == 2

# Konversi pemenangnya berupa pemain X / O jika True
def check_game(board):
  return any([bool(check_win(board)), check_tie(board)])

# 6e. Utility function should accept a terminal board and output the utility of the board.
def game(choice, turn, board):
  turn_choice_dict = {1 : 'X', -1 : 'O'}

  while not check_game(board):
    board_display(board)
    # Mengisi X / O sesuai input User
    if turn_choice_dict[turn] == choice:
      action = input("\n(The player's turn) Enter the empty position you want to place your '{}': ".format(choice))
      action = int(action)
    
    else:
      action = enemy_movement(board, choice, turn)

    if bool(check_movement(board, action, turn_choice_dict[turn])):
      board = check_movement(board, action, turn_choice_dict[turn])
      turn *= -1

    else:
      if turn_choice_dict[turn] == choice:
        print("Please select an empty spot and try again.")

  board_display(board)
  if type(check_win(board)) == np.str_:
    print('\n', check_win(board), 'won! Yeayy :)')
  
  else:
    print('\nMatch Result: Tie game')

def check_game_state(board, choice):
  if check_game(board):
    if type(check_win(board)) == np.str_:
      if check_win(board) == choice:
        return -1
      return 1
    return 0

# 6f. The minimax function should return optional action for the agent
def minimax_algorithm(board, choice, turn):
  if type(check_game_state(board, choice)) == int:
    return check_game_state(board, choice)
  
  turn_choice_dict = {1 : 'X', -1 : 'O'}

  flat_board = board[0] + board[1] + board[2]

  possible_moves = [int(x) for x in flat_board if x != 'O' and x != 'X']

  result = []
  for move in possible_moves:
    result.append(minimax_algorithm(check_movement(board, move, turn_choice_dict[turn]), choice, turn*-1))

  if turn_choice_dict[turn] == choice:
    return min(result)
  return max(result)

def enemy_movement(board, choice, turn):
  turn_choice_dict = {1 : 'X', -1 : 'O'}

  flat_board = board[0] + board[1] + board[2]

  possible_moves = [int(x) for x in flat_board if x != 'O' and x != 'X']

  result = []
  for move in possible_moves:
    result.append(minimax_algorithm(check_movement(board, move, turn_choice_dict[turn]), choice, turn*-1))
  
  print(result)
  return possible_moves[np.argmax(result)]
```
